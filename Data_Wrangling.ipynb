```python
#وقتی که بخواهیم ستونی با تشایه ایندکس به هم اضافه کنیم
a = pd.DataFrame([np.nan, 2.5, np.nan, 3.5, 4.5, np.nan],index=['f', 'e', 'd', 'c', 'b', 'a'], columns=['ff'])
b = pd.DataFrame(np.arange(len(a), dtype=np.float64),index=['f', 'e', 'd', 'c', 'b', 'a'], columns=['ss'])
a.join(b, how='outer')
```

```python
#وقتی بخواهیم به سطر اصافه کنیم
df1 = pd.DataFrame({'a': [1., np.nan, 5., np.nan],'b': [np.nan, 2., np.nan, 6.],'c': range(2, 18, 4)})
df2 = pd.DataFrame({'a': [5., 4., np.nan, 3., 7.],'b': [np.nan, 3., 4., 6., 8.]})
reult = pd.concat([df2, df1], keys=['first', 'last'])
reult.unstack()
```

```python
#وقتی بخواهیم ستونی با تشابه یک ستون به هم اضافخ کنیم
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],'data1': range(7)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],'data2': range(3)})
print(pd.merge(df1, df2, on='key'))
pd.merge(df2, df1, how='outer')
```
