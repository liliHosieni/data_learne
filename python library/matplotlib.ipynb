```python
import matplotlib.pyplot as plt
import numpy as np
import math 
```

```python
x = np.arange(1,10)
y = np.log(x)
plt.plot(x , y, "y-")

plt.xlabel("input")
plt.ylabel("sin")
plt.title("fuck you")
plt.legend(labels="1", loc='lower right')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y)
ax.grid(True)
plt.show()
```

```python

plt.subplot(131)
plt.plot(x, y)
plt.subplot(132)
plt.plot([1,4,34,3],[3,4,7,4])
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
a = [1,2,3,4]
b = [5,6,7,8]
ax.plot(a, b)
ax.set_xticks([1,2,3,4])
ax.set_xticklabels(['a','b','c','f'])
plt.show()
```

```python
x = np.arange(1,10)
y = np.arange(11, 20)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(x, y)
plt.show()
```

```python
x = np.arange(1,10)
y = np.arange(11, 20)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.hist(x,y)
plt.show()
```

```python
x = [20, 30, 40, 10]
y = ["f","e",'t','o']
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.pie(x, labels=y)
plt.show()
```

```python
x = np.arange(1,10)
y = np.arange(11, 20)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.scatter(x, y)
plt.show()
```
